```python
import kagglehub
import os
import pandas as pd
from datetime import datetime

# === 1) Cria pasta destino se não existir ===
RAW_PATH = "../dados/raw/"
os.makedirs(RAW_PATH, exist_ok=True)

# === 2) Baixa o dataset automaticamente do Kaggle ===
print("Baixando o dataset 'The Movies Dataset' do Kaggle...")
path = kagglehub.dataset_download("rounakbanik/the-movies-dataset")
print("Download concluído em:", path)

# === 3) Lista os arquivos baixados ===
print("\nArquivos disponíveis:")
for f in os.listdir(path):
    print(" -", f)

# === 4) Lê alguns arquivos principais ===
movies = pd.read_csv(os.path.join(path, "movies_metadata.csv"), low_memory=False)
credits = pd.read_csv(os.path.join(path, "credits.csv"))
#ratings = pd.read_csv(os.path.join(path, "ratings.csv"))

print("\nDimensões das tabelas:")
print("Movies:", movies.shape)
print("Credits:", credits.shape)
#print("Ratings:", ratings.shape)

# === 5) Salva cópias dos arquivos originais na pasta dados/raw ===
print("\nSalvando cópias em dados/raw ...")
movies.to_csv(os.path.join(RAW_PATH, "movies_metadata.csv"), index=False)
credits.to_csv(os.path.join(RAW_PATH, "credits.csv"), index=False)
#ratings.to_csv(os.path.join(RAW_PATH, "ratings.csv"), index=False)
print("Arquivos copiados com sucesso!")

# === 6) Exibe amostras para validação ===
print("\nPrévia dos dados:")
display(movies.head(3))
display(credits.head(3))
#display(ratings.head(3))
```
